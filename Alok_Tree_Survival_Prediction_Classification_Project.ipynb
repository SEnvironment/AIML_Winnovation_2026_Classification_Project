# **Project Summary -**

The Tree Survival Dataset is designed to analyze and predict the survival status of tree seedlings based on various environmental, biological, and ecological factors. This dataset is well suited for a classification machine learning project, where the objective is to classify whether a tree survives or does not survive under different growing conditions. Such predictive models can help ecologists, forest managers, and environmental researchers make informed decisions regarding forest conservation, biodiversity management, and plantation planning.

The dataset contains 2,783 observations and 24 attributes, representing different characteristics of individual trees. Each record corresponds to a single tree seedling and includes information related to its species, location, soil conditions, light availability, fungal associations, and biochemical properties. The dataset combines both categorical variables (such as Species, Soil Type, Light Category, Mycorrhizal Type, and Plot information) and numerical variables (such as Light_ISF, Phenolics, Lignin, NSC, and Time), making it ideal for exploring feature engineering and classification techniques.

The primary target variable for this project is Alive, which indicates whether a tree remained alive after the observation period. In cases where the Alive column is unavailable or contains missing values, the Event variable can also be used as the target by interpreting survival events appropriately. The classification task involves predicting the survival outcome using the remaining environmental and biological features.

Several factors in the dataset contribute significantly to tree survival. Light_ISF measures the amount of light available to the tree, while Light_Cat categorizes light intensity into different levels. Soil represents different soil conditions, which influence nutrient availability and water retention. Species identifies the tree species, allowing the model to capture species-specific survival patterns. Biological variables such as AMF (Arbuscular Mycorrhizal Fungi), EMF (Ectomycorrhizal Fungi), Phenolics, Lignin, and NSC (Non-Structural Carbohydrates) provide insights into the physiological and ecological characteristics that affect growth and resilience. Additional variables such as Adult, Sterile, Conspecific, and Myco further describe the ecological interactions influencing survival.

Before building the classification model, the dataset requires preprocessing steps such as handling missing values, encoding categorical variables, scaling numerical features where necessary, and selecting relevant predictors. Exploratory Data Analysis (EDA) can be performed to understand class distribution, feature relationships, and correlations between variables. Data visualization techniques such as histograms, bar charts, heatmaps, and boxplots can help identify patterns and detect potential outliers.

Various machine learning classification algorithms can be applied to this dataset, including Logistic Regression, Decision Tree, Random Forest, Support Vector Machine (SVM), K-Nearest Neighbors (KNN), Naïve Bayes, Gradient Boosting, and XGBoost. The performance of these models can be evaluated using metrics such as Accuracy, Precision, Recall, F1-Score, Confusion Matrix, and ROC-AUC Score. Cross-validation and hyperparameter tuning can further improve model performance and generalization.

Overall, this classification project demonstrates how machine learning can be used to predict tree survival based on ecological and environmental data. By identifying the most influential factors affecting survival, the model provides valuable insights into forest ecosystem dynamics and supports evidence-based decision-making in conservation and sustainable forest management. The project also offers practical experience in data preprocessing, feature engineering, model development, evaluation, and interpretation, making it an excellent application of supervised machine learning techniques in environmental science.

### Variables Description

1. No: Seedling unique ID number.
2. Plot: Number of the field plot the seedling was planted in (1-18).
3. Subplot: Subplot within the main plot the seedling was planted in. Broken into 5 subplots (1 per corner, plus 1 in the middle) (A-E).
4. Species: Includes Acer saccharum, Prunus serotina, Quercus alba, and Quercus rubra.
5. Light ISF: Light level quantified with HemiView software. Represents the amount of light reaching each subplot at a height of 1m.
6. Light Cat: Categorical light level created by splitting the range of Light_ISF values into three bins (low, med, high).
7. Core: Year the soil core was removed from the field.
8. Soil: Species from which the soil core was taken. Includes all species, plus Acer rubrum, Populus grandidentata, and a sterilized conspecific for each species.
9. Adult: Individual tree that soil was taken from. Up to 6 adults per species. Used as a random effect in analyses.
10. Sterile: Whether the soil was sterilized or not.
Conspecific: Whether the soil was conspecific, heterospecific, or sterilized conspecific.
11. Myco: Mycorrhizal type of the seedling species (AMF or EMF).
12. SoilMyco: Mycorrhizal type of the species culturing the soil (AMF or EMF).
13. PlantDate: The date that seedlings were planted in the field pots.
14. AMF: Percent arbuscular mycorrhizal fungi colonization on the fine roots of harvested seedlings.
15. EMF: Percent ectomycorrhizal fungi colonization on the root tips of harvested seedlings.
16. Phenolics: Calculated as nmol Gallic acid equivalents per mg dry extract (see manuscript for detailed methods).
17. NSC: Calculated as percent dry mass nonstructural carbohydrates (see manuscript for detailed methods).
18. Lignin: Calculated as percent dry mass lignin (see manuscript for detailed methods).
19. Census: The census number at which time the seedling died or was harvested.
20. Time: The number of days at which time the seedling died or was harvested.
21. Event: Used for survival analysis to indicate status of each individual seedling at a given time (above)
0 = harvested or experiment ended
1 = dead
22. Harvest: Indicates whether the seedling was harvested for trait measurement.
23. Alive: Indicates if the seedling was alive at the end of the second growing season. "X" in this field indicates alive status.

````python
#Droping the null Value of Event Column.
ts.dropna(subset=['Event'], inplace=True)

#Replacing Null value of Alive to No.
ts['Alive']= ts['Alive'].fillna('No')

#Replacing Null Value of Harvest to No.
ts['Harvest']= ts['Harvest'].fillna('No')

#Replacing from null Value from the mode of EMF.
ts['EMF']= ts['EMF'].fillna(ts['EMF'].mode()[0])
````
